```julia
function multiply_test()
    c = CairoRGBSurface(1000,1000)
    ctx = CairoContext(c)
    set_source_rgb(ctx,0,0,0)
    rectangle(ctx,0,0,1000,1000)
    fill(ctx)
    set_line_width(ctx,10)
    set_source_rgb(ctx,0,0,0)
    rectangle(ctx, 300, 300, 400, 400)
    stroke(ctx)
    set_operator(ctx,18)
    set_source_rgb(ctx,1,0,0)
    move_to(ctx, 200, 800)
    line_to(ctx, 800, 200)
    stroke(ctx)
    c
end
multiply_test()
```
